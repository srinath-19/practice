```python
!pip install chromadb sentence-transformers langchain-community requests -q

import requests
import chromadb
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma

client = chromadb.Client()
try:
    client.delete_collection("langchain")
except:
    pass

url = "https://ryosuzuki.github.io/lecture-demo/sample-data/ecommerce-doc.json"
docs = requests.get(url).json()
print(f"Loaded {len(docs)} documents")

PROMPT_TEMPLATE = """Use the following pieces of context to answer the question at the end. If you don't know the answer, just say that you don't know, don't try to make up an answer.

{context}

Question: {question}
Helpful Answer:"""

embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
vectorstore = Chroma.from_texts(texts=docs, embedding=embeddings, ids=[f"doc_{i}" for i in range(len(docs))])
```

```python
def generate_prompt(question, k=5):
    relevant_docs = vectorstore.similarity_search(question, k=k)
    context = "\n\n".join([doc.page_content for doc in relevant_docs])
    return PROMPT_TEMPLATE.format(context=context, question=question)

# Example Prompts
prompts = [
    "I want to buy the Sony headphones. How much would it cost with express shipping?",
    "I want to buy the MacBook Air. Can I pay in installments and how long is the warranty?",
    "What noise-cancelling headphones do you have under $300 and do you offer student discounts?",
    "If my AirPods arrive damaged, what should I do and how long until I get a refund?",
    "I'm buying a monitor for over $500. What are my shipping options and payment methods?",
    "Do you have any portable chargers I can take on a flight and what's the return policy if I don't like it?"
]

prompt = generate_prompt(prompts[0])
print(prompt)
```
